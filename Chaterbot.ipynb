```python
# Creamos una nueva instancia de Chatbot
bot = ChatBot('Mi primer bot')
```

```python
#Entrenamiento en forma de lista
trainer = ListTrainer(bot)
#Cada elemento de la lista como una posible
#respuesta a su predecesor en la lista.
trainer.train([
 "Hi there!",
 "Hello",
])
trainer.train([
 "Greetings!",
 "Hello",
])
```
